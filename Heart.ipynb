```python
import numpy as np
import pandas as pd

# Load the CSV file into a pandas DataFrame
file_path = 'files/heart.csv'
heart_data = pd.read_csv(file_path)

# Extract the numerical data into a NumPy array
numerical_data = heart_data.select_dtypes(include=[np.number]).to_numpy()

# Display the shape and the first few rows of the NumPy array to give an idea of its structure
numerical_data_shape = numerical_data.shape
numerical_data_sample = numerical_data[:5]

numerical_data_shape, numerical_data_sample
```

```python
import matplotlib.pyplot as plt
# Create the scatter plot with enhancements
plt.figure(figsize=(10, 7))

# Add scatter points with size proportional to cholesterol levels for visual enhancement
plt.scatter(heart_data['age'], heart_data['chol'], alpha=0.6, c=heart_data['chol'], cmap='viridis', s=heart_data['chol']*0.5)

# Add a color bar for cholesterol levels
plt.colorbar(label='Cholesterol Level')

# Add titles and labels
plt.title('Enhanced Age vs Cholesterol Scatter Plot', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Cholesterol (mg/dL)', fontsize=12)

# Enhance grid lines and display plot
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
